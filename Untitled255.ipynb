```python
import pandas as pd
```

```python
#load datasets
movies_df=pd.read_csv("movies.csv")
rating_df=pd.read_csv("ratings.csv")
```

```python
movies_df.head()
```

```python
rating_df.head()
```

```python
movies_df['title'] = movies_df['title'].str.replace(r'\s*\(\d{4}\)', '', regex=True)
movies_df['title'] = movies_df['title'].str.replace(r'\s*\(\)\s*', '', regex=True).str.strip()
```

```python
movies_df.head()
```

```python
#remove | from genres
movies_df['genres']=movies_df['genres'].str.split('|')
```

```python
movies_df.head()
```

```python
rating_df.drop('timestamp',axis=1,inplace=True)
```

```python
rating_df.head()
```

```python
#one hot encoding-convert the genre categories into numbers
movies_copy=movies_df.copy()
#loop through each movie
for i,row in movies_df.iterrows():
  #incase a movie has several genres
   for genre in row['genres']:
    #create a column-check if a movie belongs to a certain a genre-will be assigned 1
    movies_copy.at[i,genre]=1
#replace the null values in movies copy
movies_copy=movies_copy.fillna(0)
```

```python
movies_copy.head()
```

```python
#user input
movie_name=input("Enter a movie:").strip()
```

```python
#find the movie name
movie=movies_df[movies_df['title'].str.contains(movie_name,case=False)]
if movie.empty:
  print("Movie not found")
else:
  #search for similar movies
  movie_id=movie.iloc[0]['movieId']
  movie_data=movies_copy[movies_copy['movieId']==movie_id]
  #remove unnecessary columns
  movie_genres=movie_data.drop(['movieId','title','genres'],axis=1)
  #compare the genres with the data
  genre_data=movies_copy.set_index('movieId')
  genre_data=genre_data.drop(['title','genres'],axis=1)
  #similarity scores
  scores=genre_data.dot(movie_genres.iloc[0])
  #sort the results
  scores=scores.sort_values(ascending=False)
  #top 10 movies
  top_ids=scores.head(10).index
  recommended_movies=movies_df[movies_df['movieId'].isin(top_ids)]
  print("Recommended Movies to:"+movie_name)
  print(recommended_movies[['title']])
```
